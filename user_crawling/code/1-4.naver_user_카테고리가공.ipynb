```python
import pandas as pd
```

음식점 카테고리 만들어둔것과 유저 크롤링한 정보를 merge

```python
res_category = pd.read_excel('./data/4_4_res_category_기준.xlsx')
res_category
```

```python
유저_데이터 = pd.read_excel('./data/4_3_naver_user_크롤링_user_df_0_끝.xlsx')
유저_데이터.head(5)
```

```python
# 데이터프레임 병합
merged_df_origin = pd.merge(유저_데이터, res_category[['업태구분', '카테고리']],
                     left_on='place_cate', right_on='업태구분', how='left')

# 새로운 컬럼 ['res_category']에 '카테고리' 값 입력
merged_df_origin['res_category'] = merged_df_origin['카테고리']

# '업태구분'과 '카테고리' 컬럼 제거 
merged_df_origin = merged_df_origin.drop(columns=['업태구분', '카테고리'])
merged_df_origin.head(5)
```

```python
# res_category 값 확인
merged_df_origin['res_category'].value_counts()
```

카테고리를 정립하지 않은 (NaN 값) 확인

```python
# 'place_cate' 열에서 NaN 값이 있는 행 추출
nan_df = merged_df_origin[merged_df_origin['res_category'].isna()]
nan_df.head(5)
```

```python
nan_df['place_cate'].value_counts()
```

NaN값만 모아서 눈으로 보고 res_category 에 다시 추가하려고 했지만, 예상보다 너무 많아서 엑셀로 변환

```python
# nan_df['place_cate'] 열만 추출하여 새로운 데이터프레임 생성하고 중복 값 제거
place_cate_df = nan_df[['place_cate']].drop_duplicates()

# 열 이름을 '업태구분'으로 변경, 카테고리는 디폴트 '없음'으로 만들기 
place_cate_df.columns = ['업태구분']
place_cate_df['카테고리'] = '없음'

# 엑셀 파일로 저장
place_cate_df.to_excel('./data/4_4_res_category_기준_추가.xlsx', index=False)

place_cate_df
```

엑셀에서 확인하고 변경할거 추가 

```python
# '업태구분'이 음식인 행의 '카테고리' 값 변경
res_category.loc[res_category['업태구분'] == '달떡볶이', '카테고리'] = '분식'
res_category.loc[res_category['업태구분'] == '한정식', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '홍차전문점', '카테고리'] = '카페'
res_category.loc[res_category['업태구분'] == '소고기', '카테고리'] = '고기'
res_category.loc[res_category['업태구분'] == '블루보틀', '카테고리'] = '카페'
res_category.loc[res_category['업태구분'] == '독일음식', '카테고리'] = '외식'
res_category.loc[res_category['업태구분'] == '한정식', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '가야밀면', '카테고리'] = '분식'
res_category.loc[res_category['업태구분'] == '진빵', '카테고리'] = '분식'
res_category.loc[res_category['업태구분'] == '가야밀면', '카테고리'] = '분식'
res_category.loc[res_category['업태구분'] == '순대,순댓국', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '101번지남산돈까스', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '닭요리', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '종합분식', '카테고리'] = '분식'
res_category.loc[res_category['업태구분'] == '닭발', '카테고리'] = '한식'
res_category.loc[res_category['업태구분'] == '닭요리', '카테고리'] = '한식'
res_category.head(10)
```

```python
# 두 데이터프레임을 연결
res_category_merge = pd.concat([place_cate_df, res_category], ignore_index=True)
res_category_merge.head(10)
```

```python
# res_category_기준 엑셀에 업데이트
res_category_merge.to_excel('./data/res_category_기준.xlsx', index=False)

# 방금 확인하느라 만든 엑셀 파일은 삭제 
import os

# 삭제할 파일 경로
file_path = './data/4_4_res_category_기준_추가.xlsx'

# 파일 존재 여부 확인 후 삭제
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"{file_path} 파일이 삭제되었습니다.")
else:
    print(f"{file_path} 파일이 존재하지 않습니다.")
```

다시 유저데이터 카테고리와 비교

```python
# 데이터프레임 병합
merge_df = pd.merge(유저_데이터, res_category_merge[['업태구분', '카테고리']],
                     left_on='place_cate', right_on='업태구분', how='left')

# 새로운 컬럼 ['res_category']에 '카테고리' 값 입력
merge_df['res_category'] = merge_df['카테고리']

# '업태구분'과 '카테고리' 컬럼 제거 
merge_df = merge_df.drop(columns=['업태구분', '카테고리'])

# 'res_category' 값이 '없음'인 행 삭제
merge_df = merge_df[merge_df['res_category'] != '없음']

merge_df.head(5)
```
